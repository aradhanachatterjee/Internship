# Question 1

You need to find following details:
A) Rank
B) Name
C) Artist
D) Upload date
E) Views

```python
url = 'https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos'
html = requests.get(url).text
soup = BeautifulSoup(html, 'html.parser')
```

```python
table = soup.find('table', class_='wikitable sortable')
```

```python
df = pd.DataFrame(columns=['Rank', 'Name', 'Uploader', 'Views (Billions)', 'Upload Date'])
```

```python
for row in table.tbody.find_all('tr'):    
    columns = row.find_all('td')
    
    if(columns != []):
        rank = columns[0].text.strip()
        name = columns[1].text.strip()
        uploader = columns[2].text.strip()
        views = columns[3].text.strip()
        upload_date = columns[4].text.strip()

        df = df.append({'Rank': rank,  'Name': name, 'Uploader': uploader, 'Views (Billions)': views, 'Upload Date': upload_date}, ignore_index=True)
```

```python
df
```

```python
table = soup.find('table', class_='in-article sortable')
```

```python
df = pd.DataFrame(columns=['Rank','Book', 'Author', 'Volumes Sold', 'Publisher', 'Genre'])
```

```python
for row in table.tbody.find_all('tr'):    
    columns = row.find_all('td')
    
    if(columns != []):
        rank = columns[0].text.strip()
        book = columns[1].text.strip()
        author = columns[2].text.strip()
        vol = columns[3].text.strip()
        pub = columns[4].text.strip()
        genre = columns[5].text.strip()

        df = df.append({'Rank': rank,  'Book': book, 'Author': author, 'Volumes Sold': vol, 'Publisher': pub, 'Genre': genre}, ignore_index=True)
```
