## Question 2

```python
product_urls = []
brand_names = []
product_names = []
prices = []
returns = []
delivery = []
availability = []
a = 0
```

```python
for i in range(3):
    a+= 1
    # fetching urls for each product
    for i in d.find_elements_by_xpath('//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]'):
        product_urls.append(i.get_attribute('href'))
    for i in product_urls:
        d.get(i)
        # fetching product names
        product_name = d.find_element_by_xpath('//h1[@class="a-size-large a-spacing-none"]').text
        product_names.append(product_name)
        # fetching brand names
        try: 
            brand_name = d.find_element_by_xpath('//td[@class="a-span9"]').text
            brand_names.append(brand_name)
        except NoSuchElementException:
            brand_names.append('-')
        # fetching prices
        try:
            price = d.find_element_by_xpath('//span[@class="a-price-whole"]').text.replace('₹','')
            prices.append(price)
            availability.append('Available')
        except NoSuchElementException:
            prices.append('-')
            availability.append('Out of Stock')
        # fetching returns
        try:
            return_policy = d.find_element_by_xpath('//div[@class="a-column a-span3 a-text-center icon-container icon-grid-width celwidget"]').text
            returns.append(return_policy)
        except NoSuchElementException:
            returns.append('-')
        # fetching delivery date
        try:
            delivery_info = d.find_element_by_xpath('//div[@class="a-spacing-base"]/b').text
            delivery.append(delivery_info)
        except NoSuchElementException:
            delivery.append('-')
    if a == 1:
        d.get('https://www.amazon.in/s?k=guitar&crid=24BV87T07BFEC&sprefix=guitar%2Caps%2C199&ref=nb_sb_noss_2')
    else:
        d.get('https://www.amazon.in/s?k=guitar&page=2&crid=24BV87T07BFEC&qid=1653221225&sprefix=guitar%2Caps%2C199&ref=sr_pg_1')
    d.find_element_by_xpath('//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]').click()
```

```python
df = pd.DataFrame({'Name of product': product_names[0:116], 'Brand':brand_names[0:116], 'Price': prices[0:116], 'Return/Exchange':returns[0:116], 'Expected Delivery': delivery[0:116], 'Availability': availability[0:116], 'Product URL': product_urls})
```

```python
df['Return/Exchange'] = df['Return/Exchange'].replace('Pay on Delivery', '7 Days Replacement')
```

```python
df
```

```python
df.to_csv('amazon guitars.csv')
```

```python
d.close()
```
