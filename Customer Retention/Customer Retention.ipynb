## Effect of gender on website preference

```python
sns.countplot(df['gender'])
plt.xticks(rotation=90)
```

```python
df.gender.value_counts()
```

```python
male_preference = df2.loc[df2.gender == 0, 'user_choice']
sns.countplot(male_preference)
plt.xticks(rotation=90)
plt.title('Male Preference of E-Retail Websites')
```

```python
female_preference = df2.loc[df2.gender == 1, 'user_choice']
sns.countplot(female_preference)
plt.xticks(rotation=90)
plt.title('Female Preference of E-Retail Websites')
```

```python
plt.figure(figsize=(10,10))
sns.violinplot(x='gender', y='user_choice', data=df2)
```
