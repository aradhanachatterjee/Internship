# Question 1

```python
# Opening chrome window and going to the website
driver = webdriver.Chrome('chromedriver.exe')
driver.get('https://www.naukri.com/')

# entering 'data analyst' in job field
search_field_designation = driver.find_element_by_class_name('suggestor-input')
search_field_designation.send_keys('Data Analyst')

# entering 'Bangalore' in location field
search_field_location = driver.find_element_by_xpath('/html/body/div/div[2]/div[3]/div/div/div[3]/div/div/div/input')
search_field_location.send_keys('Bangalore')

# clicking search button
search_button = driver.find_element_by_xpath('/html/body/div/div[2]/div[3]/div/div/div[6]')
search_button.click()
```

```python
# adding job titles to list
job_titles = [i.text for i in driver.find_elements_by_xpath('//a[@class="title fw500 ellipsis"]')]

# adding locations to list
locations = [i.text.strip() for i in driver.find_elements_by_xpath('//li[@class="fleft grey-text br2 placeHolderLi location"]')]

# adding company names to list
companies = [i.text for i in driver.find_elements_by_xpath('//a[@class="subTitle ellipsis fleft"]')]

# adding experience required to list
experience = [i.text for i in driver.find_elements_by_xpath('//li[@class="fleft grey-text br2 placeHolderLi experience"]/span')]

# creating a dictionary with the scraped data and selecting top 10 out of the lists
dict1 = {'Job Title': job_titles[0:10], 'Location': locations[0:10], 'Company': companies[0:10], 'Experience Required': experience[0:10]}

# Creating dataframe using the dictionary
df = pd.DataFrame(dict1)
df
```

```python
# closing the driver window
driver.close()
```

# Question 4

```python
# Opening chrome window and going to the website
driver = webdriver.Chrome('chromedriver.exe')
driver.get('https://www.flipkart.com/')

# close pop-up
close_popup = driver.find_element_by_xpath('/html/body/div[2]/div/div/button')
close_popup.click()

# entering 'sunglasses' in product field
search_field = driver.find_element_by_xpath('/html/body/div[1]/div/div[1]/div[1]/div[2]/div[2]/form/div/div/input')
search_field.send_keys('Sunglasses')

# clicking search button
search_button = driver.find_element_by_xpath('/html/body/div[1]/div/div[1]/div[1]/div[2]/div[2]/form/div/button')
search_button.click()
```

```python
# creating empty lists to add scraped data to
brand = []
product_description = []
prices = []
discounts = []
```

```python
# Scraping data off the first 3 pages
a=11
for i in range(3):
    # getting brand name
    for i in driver.find_elements_by_xpath('//div[@class="_2WkVRV"]'):
        brand.append(i.text)
    # getting product description
    for i in driver.find_elements_by_xpath('//a[@class="IRpwTa"]'):
        product_description.append(i.text)
    # getting prices
    for i in driver.find_elements_by_xpath('//div[@class="_30jeq3"]'):
        prices.append(i.text)
    # getting discounts
    for i in driver.find_elements_by_xpath('//div[@class="_3Ay6Sb"]'):
        discounts.append(i.text)
    #clicking next button
    next_button = driver.find_element_by_xpath(f'/html/body/div[1]/div/div[3]/div[1]/div[2]/div[12]/div/div/nav/a[{a}]/span')
    next_button.click()
    if a % 2 != 0:
        a += 1
    #using time to pause for 2 seconds
    time.sleep(2)
```

```python
# creating a dictionary with the scraped data and selecting top 100 out of the lists
dict1 = {'Brand': brand[0:100], 'Product Description': product_description[0:100], 'Price': prices[0:100], 'Discount': discounts[0:100]}

# Creating dataframe using the dictionary
df = pd.DataFrame(dict1)
df
```

```python
# closing the driver window
driver.close()
```

# Question 5

```python
# Opening chrome window and going to the website
driver = webdriver.Chrome('chromedriver.exe')
driver.get('https://www.flipkart.com/apple-iphone-11-black-64-gb-includes-%20earpods-power-%20adapter/p/itm0f37c2240b217?pid=MOBFKCTSVZAXUHGR&lid=LSTMOBFKC%20TSVZAXUHGREPBFGI&marketplace.')
```

```python
# opening reviews page
reviews_page = driver.find_element_by_xpath('/html/body/div[1]/div/div[3]/div[1]/div[2]/div[9]/div/div/div[5]/div/a/div/span')
reviews_page.click()
```

```python
# creating empty lists to add scraped data
rating = []
review_summary = []
full_review = []
a=11
```

```python
# creating a loop to go through 10 pages of reviews and scraping 100 reviews
for i in range(10):
    # getting rating
    for i in driver.find_elements_by_xpath('//div[@class="_3LWZlK _1BLPMq"]'):
        rating.append(i.text)
    # negative reviews have different class
    for i in driver.find_elements_by_xpath('//div[@class="_3LWZlK _1rdVr6 _1BLPMq"]'):
        rating.append(i.text)
    # getting review summary
    for i in driver.find_elements_by_xpath('//p[@class="_2-N8zT"]'):
        review_summary.append(i.text)
    # getting full summary
    for i in driver.find_elements_by_xpath('//div[@class="t-ZTKy"]'):
        full_review.append(i.text.replace('\n\n','').replace('\n',''))
    # clicking next button
    next_button = driver.find_element_by_xpath(f'/html/body/div/div/div[3]/div/div/div[2]/div[13]/div/div/nav/a[{a}]/span')
    next_button.click()
    if a % 2 != 0:
        a += 1
    # using time to pause for 1 second
    time.sleep(1)
```

```python
# checking length of the lists
print(len(rating),len(review_summary),len(full_review))
```

```python
# creating dataframe
df = pd.DataFrame({'Rating': rating, 'Review Summary': review_summary, 'Full Review': full_review})
df
```

```python
# closing the driver window
driver.close()
```

# Question 7

```python
# Opening chrome window and going to the website
driver = webdriver.Chrome('chromedriver.exe')
driver.get('https://www.myntra.com/shoes')
```

```python
# selecting price Rs. 7074 to Rs. 14049
price_selector = driver.find_element_by_xpath('/html/body/div[2]/div/div[1]/main/div[3]/div[1]/section/div/div[5]/ul/li[2]/label/div')
price_selector.click()
```

```python
# selecting colour black
colour_selector = driver.find_element_by_xpath('/html/body/div[2]/div/div[1]/main/div[3]/div[1]/section/div/div[6]/ul/li[1]/label/div')
colour_selector.click()
```

```python
# Creating lists to add scraped data
brand = []
shoe_description = []
prices = []
```

```python
# Scraping data off the first 2 pages
for i in range(2):
    # getting brand name
    for i in driver.find_elements_by_xpath('//h3[@class="product-brand"]'):
        brand.append(i.text)
    # getting shoe description
    for i in driver.find_elements_by_xpath('//h4[@class="product-product"]'):
        shoe_description.append(i.text)
    # getting prices
    price_1 = []
    for i in driver.find_elements_by_xpath('//div[@class="product-price"]/span'):
        price_1.append(i.text)
    for i in price_1:
        if '%' not in i:
            prices.append(i[0:9].replace('R','').replace('s. ',''))
    #clicking next button
    next_button = driver.find_element_by_xpath(f'/html/body/div[2]/div/div[1]/main/div[3]/div[2]/div/div[2]/section/div[2]/ul/li[12]/a')
    next_button.click()
    #using time to pause for 2 seconds
    time.sleep(2)
```

```python
# checking length of each list
print(len(brand), len(shoe_description), len(prices))
```

```python
# creating dataframe
df = pd.DataFrame({'Brand': brand, 'Shoe Description': shoe_description, 'Price': prices})
df
```

```python
# closing the driver window
driver.close()
```

```python
# Opening chrome window and going to the website
driver = webdriver.Chrome('chromedriver.exe')
driver.get('https://www.ambitionbox.com/')

# going to salaries page
driver.find_element_by_xpath('/html/body/div[1]/nav/nav/a[4]').click()
```

```python
# entering data scientist in salary comparison field
driver.find_element_by_xpath('/html/body/div/div/div/main/section[1]/div[2]/div[1]/span/input').send_keys('Data Scientist')
```

```python
# click on data scientist
driver.find_element_by_xpath('/html/body/div/div/div/main/section[1]/div[2]/div[1]/span/div/div/div[1]/div/div').click()
```

```python
company = [i.text for i in driver.find_elements_by_xpath('//div[@class="name"]/a')]
```

```python
salary_record = [i.text.replace('based on ','').replace('salaries', '') for i in driver.find_elements_by_xpath('//div[@class="name"]/span')]
```

```python
maxmin_salary = [i.text for i in driver.find_elements_by_xpath('//div[@class="value body-medium"]')]
```

```python
minimum_salary = [maxmin_salary[i] for i in range(len(maxmin_salary)) if i % 2 == 0]
```

```python
maximum_salary = [maxmin_salary[i] for i in range(len(maxmin_salary)) if i % 2 != 0]
```

```python
average_salary = [i.text for i in driver.find_elements_by_xpath('//p[@class="averageCtc"]')]
```

```python
experience = [i.text[19:].replace(' exp','') for i in driver.find_elements_by_xpath('//div[@class="salaries sbold-list-header"]')]
```

```python
df = pd.DataFrame({'Company': company, 'Total Salary Record': salary_record, 'Average Salary': average_salary, 'Minimum Salary': minimum_salary, 'Maximum Salary': maximum_salary, 'Experience Required': experience})
df
```
